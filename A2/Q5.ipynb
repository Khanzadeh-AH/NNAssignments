```python
import numpy as np
from scipy.interpolate import Rbf
import matplotlib.pyplot as plt
```

```python
def convert_degree_to_radian(degree):
    return degree * np.pi / 180
```

```python
x = np.arange(-4, 5, 1)
y = np.array([np.sin(convert_degree_to_radian(i)) for i in x])
```

```python
interpol_func = Rbf(x, y, function="gaussian")
y_pred = interpol_func(x)
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, y, color="blue")
ax[0].set_title("Sin", color="red")
ax[1].plot(x, y_pred, color="green")
ax[1].set_title("RBF interpolated func", color="red")
fig.tight_layout()
```
