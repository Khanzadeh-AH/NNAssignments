```python
import numpy as np
from scipy.interpolate import Rbf
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 100)
y = np.array([np.sin(i * np.pi) for i in x])
```

```python
interpol_func = Rbf(x, y, function="gaussian")
y_pred = interpol_func(x)
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, y, marker="o" ,color="blue")
ax[0].set_title("Sin", color="purple")
ax[1].plot(x, y_pred, marker="o", color="red")
ax[1].set_title("RBF interpolated func", color="purple")
fig.tight_layout()
```

```python
plt.plot(x, y, color="blue", marker="o", label="sin")
plt.plot(x, y_pred, color="red", marker="o", label="RBF")
plt.legend()
plt.show()
```
